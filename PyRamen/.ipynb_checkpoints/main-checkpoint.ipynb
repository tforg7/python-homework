```python
# Import all the libraries that will be required
import csv
from pathlib import Path


## First steps of this code will translate the information in two csv files and put it in lists of list in order to manipulate the information later
# Set up a list for the information in the menu.data.csv
menu = []

# Set up a variable with the path to the csv file menu.data.csv
menupath = Path('menu_data.csv')

with open(menupath, 'r') as csvdata:
    menureader = csv.reader(csvdata, delimiter=',' )
    
    # Read the header row first and initialize a variable to keep the data
    csv_header = next(menureader)
    
    # Read all the rows of the csv file with a for loop
    for row in menureader:
        item = row[0]
        category = row[1]
        description = row[2]
        price = float(row[3])
        cost = float(row[4])
        
        # Make a list with the content of the csv file
        record = [item , category , description , price , cost]
        
        # Add every row (with a list) in the list menu
        menu.append(record)    
        
        
# Set up a list for the information in the sales.data.csv
sales = []

# Set up a variable with the path to the csv file sales.data.csv
salespath = Path('sales_data.csv')

# Open sales.data.csv with the open() function in the library csv and 
with open(salespath, 'r') as csvdata:
    salesreader = csv.reader(csvdata, delimiter=',' )
    
    # Read the header row first and initialize a variable to keep the data
    csv_header = next(salesreader)
    
    # Read all the rows of the csv file with a for loop
    for row in salesreader:
        ID = row[0]
        date = row[1]
        credit_card = row[2]
        quantity_sold = int(row[3])
        menu_item = row[4]
        
         # Make a list with the content of the csv file
        record = [ID , date , credit_card , quantity_sold , menu_item]
        
        # Add every row (with a list) in the list sales
        sales.append(record)
        
        
        
## Second step: makes a report dictionary to keep the analyzed information

        # Initialize dictionary for the different metrics to analyze (Step not needed)
            # report_figures = {"01-count" : 0,
            #                   "02-revenue" : 0 ,
            #                   "03-cogs" : 0 ,
            #                   "04-profit" : 0}

# Initialize a nestled dictionary for each item of the menu
            #  "menu_item" : report_figures   (description not needed at this stage)
report = {}     



## Third step define a function to check if a key already exist in the dictionary report. Function return a Boolean
def checkKey(dict, key):
    if key in dict.keys():
        present = True
    else:
        present = False
    return present    
    


## Export the relavant information out of the list sales to new dictionary report
# Go over Sales (row by row) to get the sales_items and the quantity of each sales_items. And add that information in the dictionary report.
for row in sales:
    # Intitialize i as a integer with Row index
    i = sales.index(row)
    
    # Initialize a local variable quantity with the value of the unit sold at a certain moment.
    quantity = sales[i][3]
    
    # Initialize a local variable sales_item with the name of the ramen sold at a certain moment.
    sales_item = sales[i][4]
    
    # Use the fucntion checkKey to determin if sales_item is already a key in the dictionary report. Function return a Boolean
    control = checkKey(report, sales_item)
    
    # Logic that applies if sales_item is not a key in the dictionary report
    if control == False:
        # Make a dictionary similar to the format in report, but with no value 
        record = {"01-count" : 0,
                  "02-revenue" : 0 ,
                  "03-cogs" : 0 ,
                  "04-profit" : 0}
        # Add new key in the dictionary report
        report[sales_item] = record 
    
    # Compute quantity values for every Key on dictionary report
    report[sales_item]["01-count"] += quantity


## Compute the Revenue, cogs and profit with information in the list menu and save it in the dictionary report    
# check in the menu list what is the price and cost for a given menu_item
for row in report:
    quantity = report[row]["01-count"]
    
    # check in the menu list what is the price and cost for a given menu_item
    for line in menu:
        
        # Intitialize l as a integer for with line index
        l = menu.index(line)
        
        # Initialize a local variable with the value of the price for a specific menu_item
        price = menu[l][3]
        
        # Initialize a local variable with the value of the cost for a specific menu_item
        cost = menu[l][4]
        
        # calculate the profit based on the price and the cost
        profit = price - cost
        
        #If the name of the key for a row is the same as the menu_item of the list menu, then the values of price and cost are used
        if row == menu[l][0]:
            
            # Values in the dictionary are adapted 
            report[row]["02-revenue"] += (price * quantity)
            report[row]["03-cogs"] += (cost * quantity)
            report[row]["04-profit"] += (profit * quantity)
             

# create a new file called report.txt in the same directory
output = Path("report.txt")

# Open report.txt
with open(output, 'w') as texte:
    
    # Write every line of the dictionary report in the file report.txt
    for key, value in report.items():
        
        # Write line of dictionary with the layout " key :value " on one line and then break line
        texte.write('%s:%s\n' % (key, value))

```
