```python

# Initialize metric variables
total_month = 0
total = 0
Average_change = 0

# Initialize Market evolution variables : maximum key-value pair
Greatest_Increase = 0
Greatest_Increase_date = "Not determined"

# Initialize Market evolution variables : minimum key-value pair
Greatest_Decrease = 0
Greatest_Decrease_date = "Not determined"

# Initialize memory variables necessary helpfull for calculation
previous_value = 0
first_value = 0

# Import Pathlib library and csv library to access and read the data document.
from pathlib import Path
import csv

# Initialize path to csv file with the fonction path()
csvpath = Path('budget_data.csv')

# Initialize a dictionary file with which we can work to compute the market analysis.
dict_dat = {}

# Open csv file and read it using the function .reader() from the csv library
with open(csvpath, 'r') as csvdata:
    csvreader = csv.reader(csvdata, delimiter=',')
    
    # Read the header row first and initialize a variable to keep the information. Skipping the header allows to keep only the values in the for loop.
    csv_header = next(csvreader)
    
    # Iterate the input of key-values in the dictionnary 'dict-data' with the data coming from csvreader.
    # N.B.: The header is not an element of this dictionnary as it has been next() in the section above
    dict_data = {rows[0]:rows[1] for rows in csvreader}
   
    # Iterate over key-value pairs of the dictionary
    for  Date, Profit in dict_data.items() :
        
        # Count the number of times that the for loop is computing. This is equivalent to the number of months.
        total_month += 1
        
        # Makes a total sum of all the values in the dictonnary
        total += int(Profit)
        
        # Logic to determine the increase between two consecutive months. This first logic is used to make a specific case for the first iteration.
        if previous_value == 0 :
            current_increase = 0
            first_value = int(Profit)
        
        # This part of the logic takes all cases except the first row of the dictonnary
        else :
            current_increase = int(Profit) - previous_value
        
        # Logic process to determine the greatest increase and decrease between two consecutive months. 
        # This difference is calcultate between the current month and the previous month.
            if current_increase > Greatest_Increase :
                Greatest_Increase = current_increase
                Greatest_Increase_date = Date
            
            elif current_increase < Greatest_Decrease :
                Greatest_Decrease = current_increase
                Greatest_Decrease_date = Date
            
            # calculate and round the average change between consecutive months by taking the first and the last value and dividing it by the amont of months minus 1
            Average_change = round((int(Profit) - first_value) / (total_month-1) , 2)
        
        # Allocation of a new value at the end of the logic steps to keep the previous value in the next iteration.
        previous_value = int(Profit)

# Set the path for the output.csv
output = Path("output.txt")

# Open the output path as a file
f = open(output, 'w')

# Write a text line per line in the output.txt file
f.write(" Financial Analysis:") 
f.write("\n ----------------------------")    
f.write(f"\n Total Months: {total_month}") 
f.write(f"\n Total: ${total}")    
f.write(f"\n Average  Change: ${Average_change}")  
f.write(f"\n Greatest Increase in Profits: {Greatest_Increase_date} (${Greatest_Increase})")  
f.write(f"\n Greatest Decrease in Profits: {Greatest_Decrease_date} (${Greatest_Decrease})") 
f.close()
```
